# First preprocessing of the videos

- Rezise
- FPS
- Codec


```python
import numpy as np
import os
import cv2


mainPath = f"/media/pc/backup1/BaseDeDatos/TransLowNet/Videos"

outputPath = f"/media/pc/backup1/BaseDeDatos/TransLowNet/VideosEditing"

fullType = sorted(os.listdir(mainPath))

new_width = 640
new_height = 360
new_fps = 30.0
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

for VehicleName in fullType:
    pathwayVideos = os.path.join(mainPath, f"{VehicleName}")
    fullNameVideo = sorted(os.listdir(pathwayVideos))

    outputPath_Vehicle = os.path.join(outputPath, f"{VehicleName}")
    os.makedirs(outputPath_Vehicle, exist_ok=True)


    for nameVideo in fullNameVideo:
        videoPath = os.path.join(pathwayVideos, f"{nameVideo}")
        outputPath_Video = os.path.join(outputPath_Vehicle, f"{VehicleName}_{nameVideo}")
        cap = cv2.VideoCapture(videoPath)
        if not cap.isOpened():
            print(f"Error al abrir el video: {videoPath}")
            break

        out = cv2.VideoWriter(outputPath_Video, fourcc, new_fps, (new_width, new_height))

        print(f"Procesando: {os.path.basename(videoPath)}...")

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # 4. Redimensionar el frame
            frame_resized = cv2.resize(frame, (new_width, new_height), interpolation=cv2.INTER_AREA)

            # 5. Escribir el frame en el nuevo archivo
            out.write(frame_resized)

        # Liberar recursos
        cap.release()
        out.release()
        print(f"Video guardado exitosamente en: {outputPath_Video}")


```

# Second process, frame extraction

```python
import numpy as np
import os 


mainPath = f"/media/pc/backup1/BaseDeDatos/TransLowNet/VideosEditing"
mainPath_Frames = f"/media/pc/backup1/BaseDeDatos/TransLowNet"
fullType = sorted(os.listdir(mainPath))



for VehicleName in fullType:
    pathwayVideos = os.path.join(mainPath, f"{VehicleName}")
    fullNameVideo = sorted(os.listdir(pathwayVideos))
    
    
    for nameVideo in fullNameVideo:
        outputPath_Vehicle = os.path.join(mainPath_Frames, f"dataframes" ,f"{VehicleName}", f"{nameVideo[:-4]}")
        os.makedirs(outputPath_Vehicle, exist_ok=True)
        videoPath = os.path.join(pathwayVideos, f"{nameVideo}")
        outputPath_Video = os.path.join(outputPath_Vehicle, f"{VehicleName}_{nameVideo}")
        cap = cv2.VideoCapture(videoPath)
        if not cap.isOpened():
            print(f"Error al abrir el video: {videoPath}")
            break
        countVideo = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frameName = os.path.join(outputPath_Vehicle, f"imag{countVideo:07d}.jpg")
            cv2.imwrite(frameName, frame)
            countVideo +=1
        
        cap.release()
        out.release()
        print(f"Frames del video {VehicleName}/{nameVideo[:-4]} extraidos.")
```
